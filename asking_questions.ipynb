```python
import numpy as np
import pandas as pd
```

```python
from openai import OpenAI
```

```python
client = OpenAI(
    base_url="http://192.168.15.182:1234/v1",
    api_key="sk-no-key-required"
)
```

```python
response = client.chat.completions.create(
    model="qwen/qwen3.5-35b-a3b",
    messages=[
        {"role": "user", "content": "Привет! Напиши одно короткое приветствие."}
    ]
)
```

```python
print(response.choices[0].message.content)
```

```python
usage = response.usage
```

```python
print(f"Токенов в запросе (Prompt): {usage.prompt_tokens}")
print(f"Токенов в ответе (Completion): {usage.completion_tokens}")
print(f"Всего потрачено токенов (Total): {usage.total_tokens}")
```

```python
try:
    response = client.embeddings.create(
        model="qwen3.5-397b-a17b",
        input="Проверка embedding endpoint"
    )

    embedding = response.data[0].embedding

    print("Embedding endpoint works")
    print("Vector length:", len(embedding))
    print("First 5 values:", embedding[:5])

except Exception as e:
    print("Embedding endpoint failed")
    print(type(e).__name__)
    print(e)
```
